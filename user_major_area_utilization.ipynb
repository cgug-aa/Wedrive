- 특정 사용자의 드문 활동 구역을 주된 활동 구역으로 갖는 사용자 리스트 추출하는 코드

```python
import os
from dotenv import load_dotenv
from pymongo import MongoClient
from collections import defaultdict
```

```python
# 데이터베이스 연결
load_dotenv()
client = MongoClient(os.getenv('DB_ADR'),
          username=os.getenv('DB_USER'),
          password=os.getenv('DB_PASSWORD'),
          authSource=os.getenv('DB_AuthSource'),
          authMechanism=os.getenv('DB_AuthMechanism'))
db = client.get_database(os.getenv('DB_Collection'))

#컬렉션 불러오기
cl_user_frequency_zone=db.get_collection("user_frequency_zone")
```

```python
# user_frequency_zone에서 basic-unit에 따라 uuid 딕셔너리로 정리
dict_freq=defaultdict(list)
for doc in cl_user_frequency_zone.find({},{"uuid":1, "basic-unit":1, "_id":0}):
    dict_freq[doc['basic-unit']].append(doc['uuid'])
dict_freq
```
